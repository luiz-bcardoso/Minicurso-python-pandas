```python
import pandas as pd
arquivo = '/workspaces/dev/glicose_data_suja.csv'
```

Abrindo arquivo csv e convertendo para data_frame

```python
data_frame_csv = pd.read_csv(arquivo)
data_frame_csv
```

Mostrando apenas as colunas 'Dia semana' e 'Resultado'

```python
data_frame_csv[['Dia Semana','Resultado']]
```

Mostra da linha 5 até 11 e da coluna 0 até 3.

```python
data_frame_csv.iloc[5:11,0:3]
```

Mostra todos os dados da linha 1.

> .loc vs .iloc, o iloc considera a primeira linha não sendo o nome das colunas, apenas dado.

```python
linha_0 = data_frame_csv.iloc[1]
linha_0
```

Modifica o valor do resultado na linha 3

```python
data_frame_csv.at[3,'Resultado'] = 100
data_frame_csv.iloc[3]
```

Modifica o nome da coluna 'Dia Semana' para 'Dia da Semana'

```python
data_frame_csv = data_frame_csv.rename(columns={'Dia Semana' : 'Dia da Semana'})
data_frame_csv.iloc[0]
```

Ordena a tabela pela menor data e depois menor resultado (duplamente ordenado)

```python
data_frame_ordenado_colunas = data_frame_csv.sort_values(by=['Dia da Semana', 'Resultado'], ascending=True)
data_frame_ordenado_colunas[:10]
```

Remove as linhas em que o resultado teve valor zero.

```python
data_frame_csv['Resultado'] = data_frame_csv['Resultado'].fillna(0)
data_frame_csv['Resultado']
```

Faz uma correlação entre os resultados e calorias

> Semelhante a regra de tres, onde 1und. insulina é equivalente a 0.002 kcal

```python
df_caloria_resultado = data_frame_csv[['Resultado','kcal']]
correlacao = df_caloria_resultado.corr()
correlacao
```

Define uma função que classifica o nivel de insulina com base no resultado.

```python
def classifica_resultado(valor):
    if valor <= 90:
        return 'baixo'
    elif valor <= 120:
        return 'normal'
    else:
        return 'alto'
```

Cria uma nova coluna no df passando o resultado como parametro na função de classificação

```python
df_caloria_resultado['Classifica Resultado'] = df_caloria_resultado['Resultado'].apply(classifica_resultado)
df_caloria_resultado
```

```python
df_caloria_resultado = df_caloria_resultado.sort_values(by=['Resultado'], ascending=False)
df_caloria_resultado.iloc[:10]
```

Pega todos os dados em que a classificação dos resultados foram altos

```python
df_resultados_altos = data_frame_csv.loc[df_caloria_resultado['Classifica Resultado'] == 'alto']
df_resultados_altos
```

Cria um arquivo .csv com um data_frame modificado

```python
data_frame_csv.to_csv('../glicose_data_limpa.csv', index=False)
```
